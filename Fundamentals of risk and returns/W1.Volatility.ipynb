# Volatility - Standard Deviation and Variance

```python
import pandas as pd
prices = pd.read_csv('MyOwn/AAPL.csv')
# a file containing the daily prices of two stocks (Apple Inc. and AMAZON Inc.) for a year (4/30/2020 to 4/30/2021)
returns = prices.pct_change()
returns
```

```python
returns = returns.dropna()
# this way the NA values are discarded
returns
```

```python
returns.std() # uses a denomenator which is n-1
# Calculated the standard deviation of a sample of values
```

```python
# Alternatively:

deviations = returns - returns.mean()
squared_deviations = deviations**2
variance = squared_deviations.mean() # uses a denomenator which is n

import numpy as np
volatility = np.sqrt(variance)
volatility
```

```python
# The difference between the returns.std() and the above formula is the denomenator
```

```python
# number_observations = returns.shape[0]
returns.shape
```

```python
volatility
```

```python
# annualizied volatility
volatility*np.sqrt(252)
# Since our data is based on daily prices and returns we use number of working days (252)
```

# Example

```python
returns = pd.read_csv('MyOwn/Portfolios_Formed_on_E-P.CSV',
                    header=0,index_col=0,parse_dates=True,na_values=-99.99)
# This data contain the returns on portfolios formed on E/P
returns
```

```python
returns.head()
```

```python
returns = returns[['Lo 10', 'Hi 10']]/100
returns
```

```python
returns.columns = ['Low E/P', 'High E/P']
returns
```

```python
returns.plot.line()
```

```python
returns.std()
```

```python
annualized_vol = returns.std()*np.sqrt(12)
# Here we have monthly data
annualized_vol
```

```python
num_month = returns.shape[0]
return_per_month = (1 + returns).prod()**(1/num_month) - 1
return_per_month
```

```python
annualized_vol
```

```python
annualized_return = (return_per_month+1)**12 -1  # return_per_month = (1 + returns).prod()**(12/num_month) - 1
annualized_return
```

# Sharpe Ratio

```python
rfr = 0.03 # Risk Free Rate (Just a wild guess)
Sharpe_ratio = (annualized_return - rfr)/annualized_vol
Sharpe_ratio.round(2)
```

```python
# As it is obvious, portfolios containing higher E/P stocks have a higher excess return compared to lower E/P stocks.
```
